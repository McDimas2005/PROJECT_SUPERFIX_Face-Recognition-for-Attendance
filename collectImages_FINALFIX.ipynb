```python
# 500 IMAGES in a row\
    
import cv2
import os

# Create a directory for saving images
output_dir = "DATASET/me"
os.makedirs(output_dir, exist_ok=True)

# Initialize the webcam
cap = cv2.VideoCapture(0)

# Ensure the webcam is opened correctly
if not cap.isOpened():
    print("Error: Could not access the webcam.")
    exit()

print("Capturing 500 images. Press 'q' to quit early.")

count = 0
while count < 500:
    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to capture frame.")
        break

    # Display the frame
    cv2.imshow("Capture Images", frame)

    # Save the captured image
    img_path = os.path.join(output_dir, f"me_{count}.jpg")
    cv2.imwrite(img_path, frame)
    print(f"Saved {img_path}")

    count += 1
    cv2.waitKey(100)  # wait for 100 ms between captures

# Release the webcam and close the window
cap.release()
cv2.destroyAllWindows()

print("Image capture complete. 500 images saved.")
```

```python
import os
import shutil
import random
```

```python
source_folder = r"C:\Users\Michael Dimas\Documents\BINUS\Course\Semester 5\Computer Vision\PROJECT\data\negative"
destination_folder = r"DATASET/not_me"
os.makedirs(destination_folder, exist_ok=True)
```

```python
# Number of images to select
num_images = 10000

# List all image files in the source folder
all_images = [f for f in os.listdir(source_folder) if f.endswith(('.jpg', '.jpeg', '.png'))]

# Randomly select images
selected_images = random.sample(all_images, min(num_images, len(all_images)))

# Copy selected images to the destination folder
for image in selected_images:
    source_path = os.path.join(source_folder, image)
    destination_path = os.path.join(destination_folder, image)
    shutil.copy(source_path, destination_path)

print(f"Copied {len(selected_images)} images to {destination_folder}")
```
